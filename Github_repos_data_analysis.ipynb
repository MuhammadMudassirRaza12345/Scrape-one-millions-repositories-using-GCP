```python
# Count the number of repositories for each owner
top_owners=df['Owner'].value_counts().sort_values(ascending=False).head(10)
```

```python
top_owners.plot(kind = 'bar',ylabel=("Count") ,title="Top 10 Repository Owners")
```

```python
# Identify the most popular repository names
Popular_repository_names=df['Repository Name'].value_counts().sort_values(ascending=False).head(10)
```

```python
Popular_repository_names.plot(kind = 'bar',ylabel=("Count") ,title="Top 10 Popular repository names")
```

```python
# This show most pouplar repositry name in percentage
Popular_repository_names.plot(kind = 'pie',autopct='%1.1f%%', startangle=140,title="Top 10 Popular repository names")
```

```python
df2=df.copy()
```

```python
# remove those rows whose values contains message or documentation_url
df2 = df2[~df2['Languages'].str.contains('message|documentation_url')]
```

```python
df2.tail()
```

```python
df2['Languages'] = df2['Languages'].str.split(', ')

# Create a list to store all languages
all_languages = [language for sublist in df2['Languages'].tolist() for language in sublist]

# Count the frequency of each language
language_counts = pd.Series(all_languages).value_counts()

# Analyze the most commonly used languages
most_common_languages = language_counts.head(10)

print(most_common_languages)
```

```python

# Create a bar chart
plt.figure(figsize=(10, 6))
language_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Languages in GitHub Repositories')
plt.xlabel('Languages')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python


# Assuming you have language counts in a DataFrame 'language_counts'
labels = language_counts.index[:10]  # Use the top 10 languages, for example
sizes = language_counts.head(10)
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Languages in GitHub Repositories')
plt.show()
```
